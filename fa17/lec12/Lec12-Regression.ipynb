```python
# WGP Wide 123 Dataset
WGP_Wide_123 = pd.DataFrame(pd.read_csv('WGP_wide_123.csv'))
WGP_Wide_123_Table=Table.from_df(WGP_Wide_123)
#WGP_Wide_123_Table
```

```python
#  WGP outcome variables
# these need to be transformed to be 0-1 to work with Logit or Probit
y_s2 = WGP_Wide_123['Selfrptpct_2']/100
y_v2 = WGP_Wide_123['Vldclpct_2']/100
y_s3 = WGP_Wide_123['Selfrptpct_3']/100
y_v3 = WGP_Wide_123['Vldclpct_3']/100
```

```python
# Treatment Arms
treatment = WGP_Wide_123[['treat1','treat2','treat3','treat4','treat5','treat6','treat7']]
treatment_2= WGP_Wide_123[['treat2','treat3','treat4','treat5','treat6','treat7']]
treatment_2_const =  sm.add_constant(treatment_2)
```

```python
#  possible explanatory variables ?
x1 = WGP_Wide_123['b3_age_nonan']
x2 = WGP_Wide_123['b9_education_nonan']
x3 = WGP_Wide_123['durable_nonan']
x4 = WGP_Wide_123[['b3_age_nonan','b9_education_nonan','durable_nonan']]
#  addint in a column of ones to be the constant
x1_const = sm.add_constant(x1)
x2_const = sm.add_constant(x2)
x3_const = sm.add_constant(x3)
x4_const = sm.add_constant(x4)


```

```python
#WGP_Wide_123['Selfrptpct_2']
```

```python
# Regression of self reports in round 2 vs treatment arms
regress=sm.OLS(y_s2,treatment).fit()
regress.summary()
```

```python
# Regression of validated  in round 3 vs treatment arms
regress=sm.OLS(y_v3,treatment).fit()
regress.summary()
```

```python
# Regression of validated  in round 3 vs  6 treatment arms and a constant
regress=sm.OLS(y_v3,treatment_2_const).fit()
regress.summary()
```

```python
# regression with no intercept
#regress=sm.OLS(Outcome_Diarrhea,Treatment_Spring).fit()
#regress.summary()
```

```python
# Adding intercept
Treatment_Spring_const = sm.add_constant(Treatment_Spring)
# Regression of Diarrhea vs Spring Treatment
regress=sm.OLS(Outcome_Diarrhea,Treatment_Spring_const).fit()
regress.summary()
```

```python
# Adding intercept
Treatment_WG_const = sm.add_constant(Treatment_WG)
# Regression of Diarrhea vs Spring Treatment
regress=sm.OLS(Outcome_Diarrhea,Treatment_WG_const).fit()
regress.summary()
```

```python
# Lets tryout some of the background variables on who uses WG
regress=sm.OLS(y_v2,x1_const).fit()
regress.summary()
```

```python
# Lets check the interaction of all the background variables on who uses WG
regress=sm.OLS(y_v2,x4_const).fit()
regress.summary()
```
